```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test)=datasets.cifar10.load_data()
x_train.shape
x_test.shape
```

```python
x_train.shape
```

```python
y_train=y_train.reshape(-1)
y_train[:5]
```

```python
classes=['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
def plot_sample(x, y, index):
    plt.figure(figsize=(15,2))
    plt.imshow(x[index])
    plt.xlabel(classes[y[index]])
```

```python
 plot_sample(x_train, y_train, 0)
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
netA=models.Sequential([layers.Flatten(input_shape=(32,32, 3)), layers.Dense(10, activation='sigmoid')])
netA.compile(optimizer='SGD', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
A=netA.fit(x_train, y_train, epochs=50, validation_split=0.25)
```

```python
netA.evaluate(x_test, y_test)
```

```python
netB=models.Sequential([layers.Flatten(input_shape=(32,32, 3)), layers.Dense(300, activation='relu'), layers.Dense(10, activation='sigmoid')])
netB.compile(optimizer='SGD', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
B=netB.fit(x_train, y_train, epochs=50, validation_split=0.25)
```

```python
netB.evaluate(x_test, y_test)
```

```python
netC= models.Sequential([layers.Conv2D(filters=32, kernel_size=(25,5), activation ='relu', input_shape=(32, 32, 3)),
                        layers.MaxPooling2D(pool_size=(2, 2), strides=2),
                        #layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
                        #layers.MaxPooling2D((2, 2)),
                        
                        layers.Flatten(),
                        layers.Dense(10, activation='softmax')
                       ])
```

```python
netC.compile(optimizer='adam', loss= 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
C=netC.fit(x_train, y_train, epochs=50, validation_split=0.25)
```

```python
netC.evaluate(x_test, y_test)
```

```python
nets = [netA, netB, netC]
histories=[]
```

```python
history=[A, B, C]
names=['netA', 'netB', 'netC']
i=0
for hist in history:
    fig, ax = plt.subplots(1,1,figsize=(10,4))
    plt.plot(hist.history['accuracy'], label='train_acc')
    plt.plot(hist.history['val_accuracy'], label='test_acc')
    ax.set_title(names[i])
    plt.legend()
    plt.show()
    i=i+1
```
